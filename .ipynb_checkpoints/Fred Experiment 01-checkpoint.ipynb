Navigating there I see the S&P 500 as a graph.  Terrific.

It took only a little more research in the https://fred.stlouisfed.org/docs/api/fred/ page  to figure out that I can get the S&P500 data as follows:

```python
import requests

thing = requests.get("https://api.stlouisfed.org/fred/series/observations?series_id=SP500&api_key=d242e76b770f7b3f544e85f1e00cb746&file_type=json").json()

print(thing)
```

Nice.  Now to figure out how to use the date and value fields to create a time series graph.

This video provides a example which might help: https://www.youtube.com/watch?v=_LWjaAiKaf8 

To convert thing's json to something more usable ...

```python
import datetime

def strToDateTime(s):
    sa = s.split('-')
    return datetime.date(int(sa[0]), int(sa[1]), int(sa[2]))


sp500series = [(strToDateTime(x["date"]), float(x["value"])) for x in thing["observations"] if x["value"] != "."]

print(sp500series)
```

And now set up a graph

```python
from matplotlib import pyplot as plt
from matplotlib import dates as plt_dates

plt.style.use("seaborn")

plt.plot_date([x[0] for x in sp500series], [x[1] for x in sp500series],fmt='-', linewidth=1)


plt.tight_layout()
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import dates as plt_dates
from matplotlib.patches import Circle

plt.style.use("seaborn")

fig, ax = plt.subplots()

plt.plot_date([x[0] for x in sp500series], [x[1] for x in sp500series],fmt='-', linewidth=1)

point = (datetime.date(2014, 9, 11), 1997.45)

# Draw a circle around the point
circle = Circle((point[0], point[1]), 0.5, color='red', fill=False, lw=2)
ax.add_patch(circle)

plt.tight_layout()
plt.show()
```

Now find local maxima (and circle them in red). What if two days have the same value?  Does that occur in this data set?

```python
for i in range(0, len(sp500series) - 1):
    if sp500series[i][1] == sp500series[i + 1][1]:
        print(sp500series[i])
```

It does occur. Once. So even though very infrequent, such a case can occur, so I should account for this.

```python
for i in range(0, len(sp500series) - 2):
    if sp500series[i][1] < sp500series[i + 1][1]:
        j = i + 2
        while sp500series[i + 1][1] == sp500series[j][1]: # j < len(sp500series) ?
            j += 1
        if sp500series[i + 1][1] > sp500series[j][1]:
            print(f'Found local maxima at {sp500series[i + 1]}')
```

"Matrix Convolution" might be a thing here.
