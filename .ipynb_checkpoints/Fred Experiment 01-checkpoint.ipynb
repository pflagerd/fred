Experimenting with the api at https://fred.stlouisfed.org/

First I found this link: [https://fred.stlouisfed.org/docs/api/fred/](https://fred.stlouisfed.org/docs/api/fred)

And within in it I found this link: [https://fred.stlouisfed.org/docs/api/fred/category.html](https://fred.stlouisfed.org/docs/api/fred/category.html)

This has an example, but it looks for `category_id=125`.

I want to look for the "root" category, which is `category_id=0`.

I don't remember where I found that the "root" category is 0.

```python
import requests

thing = requests.get("https://api.stlouisfed.org/fred/category?category_id=0&api_key=d242e76b770f7b3f544e85f1e00cb746&file_type=json").json()

print(thing)
```

Not very inspiring, but in conjunction with category/children, maybe more interesting?

```python
import requests

thing = requests.get("https://api.stlouisfed.org/fred/category/children?category_id=0&api_key=d242e76b770f7b3f544e85f1e00cb746&file_type=json").json()

print(thing)
```

Yes.  Now we're getting somewhere.

A little googling and I find this url: https://fred.stlouisfed.org/series/SP500

Navigating there I see the S&P 500 as a graph.  Terrific.

It took only a little more research in the https://fred.stlouisfed.org/docs/api/fred/ page  to figure out that I can get the S&P500 data as follows:

```python
import requests

thing = requests.get("https://api.stlouisfed.org/fred/series/observations?series_id=SP500&api_key=d242e76b770f7b3f544e85f1e00cb746&file_type=json").json()

print(thing)
```
