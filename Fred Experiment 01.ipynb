It took only a little more research in the https://fred.stlouisfed.org/docs/api/fred/ page  to figure out that I can get the S&P500 data as follows:

```python
import requests

thing = requests.get("https://api.stlouisfed.org/fred/series/observations?series_id=SP500&api_key=d242e76b770f7b3f544e85f1e00cb746&file_type=json").json()

print(thing)
```

Nice.  Now to figure out how to use the date and value fields to create a time series graph.

To convert thing's json to something more usable ...

```python
import datetime

def strToDateTime(s):
    sa = s.split('-')
    return datetime.date(int(sa[0]), int(sa[1]), int(sa[2]))


sp500series = [(strToDateTime(x["date"]), float(x["value"])) for x in thing["observations"] if x["value"] != "."]

print(sp500series)
```

And now set up a graph

```python
from matplotlib import pyplot as plt
from matplotlib import dates as plt_dates

plt.style.use("seaborn")

plt.plot_date([x[0] for x in sp500series], [x[1] for x in sp500series],fmt='-', linewidth=1)


plt.tight_layout()
plt.show()
```

Now find local maxima where followed by a decline of 18% or more

```python
for i in range(0, len(sp500series) - 2):
    if sp500series[i][1] < sp500series[i + 1][1] and sp500series[i + 1][1] > sp500series[i + 2][1]:
        print(f'Found local maxima at {sp500series[i + 1]}')
```
